```python
import ode
import numpy
```

```python
world = ode.World()
world.setGravity((0, -9.81, 0))
```

```python
body = ode.Body(world)
M = ode.Mass()
M.setSphere(2500.0, 0.05)
M.mass = 1.0
body.setMass(M)
```

```python
body.setPosition((0, 2, 0))
body.setLinearVel((20, 10, 0))
```

```python
total_time = 0.0
dt = 0.01
while total_time < 3.0:
    x, y, z = body.getPosition()
    u, v, w = body.getLinearVel()
    print(total_time, x, y, z)
    world.step(dt)
    total_time += dt
```
